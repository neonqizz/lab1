## Анализ и обработка данных
</div>

Colab позволяет использовать для анализа и визуализации данных все возможности популярных библиотек Python. Например, в ячейке ниже используется библиотека <strong>numpy</strong> для генерации случайных данных, а также библиотека <strong>matplotlib</strong> для их визуализации. Чтобы изменить код, достаточно нажать на ячейку.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

# Новый раздел

```
import numpy as np
data_path = "/file1.data"
data = np.genfromtxt(data_path, delimiter=",")
print(data)
```

```
print ( "Data type : ", type(data) )
print ( "Data shape : ", data.shape )
print ( data[-4:] )
```

```
data1 = np.genfromtxt(data_path, delimiter=",", dtype=None)
print('Shape of the dataset:', data1.shape)
print('Dataset type:', type(data1))
print('A single row of the dataset is type of:', type(data1[0]))
print('Types of elements:', type(data1[0][1]), type(data1[0][3]))
print('Dataset:')
print(data1)
```

```
dt = np.dtype("i1, f8, i2, U10")
data2 = np.genfromtxt(data_path, delimiter=",", dtype=dt)
print('Shape of the dataset:', data2.shape)
print('Dataset type:', type(data2))
print('A single row of the dataset is type of:', type(data2[0]))
print('Types of elements:', type(data2[0][1]), type(data2[0][3]))
print('Dataset slice:')
print(data2[:10])
```

```
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Данные из отдельных столбцов
visota = [] # Sepal Length
dlina = [] # Sepal Width
shirina = [] # Petal Length


# Выполняем обход всей коллекции data2
for dot in data2:
    visota.append(dot[0])
    dlina.append(dot[1])
    shirina.append(dot[2])

# Строим графики по проекциям данных
# Учитываем, что каждые 70 типов ирисов идут последовательно
plt.figure(1)
bomj, = plt.plot(visota[:70], dlina[:70], 'ro', label='bomj')
musor, = plt.plot(visota[70:140], dlina[70:140], 'g^', label='musor')
bootilka, = plt.plot(visota[140:210], dlina[140:210], 'bs', label='bootilka')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('visota')
plt.ylabel('dlina')

plt.figure(2)
bomj, = plt.plot(visota[:70], shirina[:70], 'ro', label='bomj')
musor, = plt.plot(visota[70:140], shirina[70:140], 'g^', label='musor')
bootilka, = plt.plot(visota[140:210], shirina[140:210], 'bs', label='bootilka')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('visota')
plt.ylabel('shirina')

plt.figure(3)
bomj, = plt.plot(shirina[:70], dlina[:70], 'ro', label='bomj')
musor, = plt.plot(shirina[70:140], dlina[70:140], 'g^', label='musor')
bootilka, = plt.plot(shirina[140:210], dlina[140:210], 'bs', label='bootilka')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('shirina')
plt.ylabel('dlina')

plt.show()
```
